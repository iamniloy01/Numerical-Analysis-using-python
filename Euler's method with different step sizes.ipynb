**_This is an initial value problem (IVP) for the first-order ordinary differential equation (ODE) dy/dx= yt^2-1.1y , y(0)=1, t=0 to 2_**

**_The ODE has an analytical solution: y=e^(t^3 /3)-1.1t). The goal is to compare the analytical solution with numerical approximations obtained by Euler's method using different step sizes h._**

**_The accuracy of Euler's method depends on the step size : smaller h produces a curve to exact solution , while larger h results in larger errors._**

**_Effect of Step Size_**

**_Large h:Few steps, faster computation, but less accurate. The numerical solution deviates from the analytical curve._**


**_Small h:More steps, slower computation, but more accurate. The numerical solution closely follows the analytical curve._**


**_Graphically, as we reduce h (e.g.,0.5,0.25,0.125,â€¦), the Euler solution curves gradually align with the analytical exponential curve._**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def f(t,y):
  return y*(t**2)-1.1*y
```

```python
def euler(ti,tf,h,y0):
  t=[ti]
  y=[y0]
  for i in np.arange(ti,tf+h,h):
    if i==0:
      continue
    else:
      y_new=y[-1]+(h*f(t[-1],y[-1]))
      t_new=t[-1]+h
      y.append(y_new)
      t.append(t_new)
  return {"t": t,"y":y}
```

```python
df2=pd.DataFrame(columns=["t","y"])
def analytical(ti,tf,h):
  for i in np.arange(ti,tf+h,h):
    y=np.exp(((i**3)/3)-1.1*i)
    df2.loc[len(df2)]=[i,y]
```

```python
analytical(0,2,.5)
```

```python
h=.5
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple', 'brown']
plt.figure(figsize=(10, 6))
plt.style.use('ggplot')
plt.plot(df2["t"], df2["y"], label="Analytical", color="black",linestyle='--') # Plot of the analytical solution
for i in np.arange(10):
  d=euler(0,2,h,1)
  plt.plot(d["t"],d["y"], label=f"h={h}", color=colors[i])
  h/=2
plt.xlabel("t",fontsize=14,color="black")
plt.ylabel("y",fontsize=14,color="black")
plt.ylabel("y")
plt.title("Euler Method with Different Step Sizes vs. Analytical Solution",color="black")
plt.legend()
plt.grid(True)
plt.show()
```

**_The black dashed line represents the analytical solution y(t)=e^(t^3 /3)âˆ’1.1t .The colored solid lines represent Eulerâ€™s method solutions with different step sizes._**

**_For large step sizes, the Euler method produces noticeable error.As step size decreases, the approximation improves, showing convergence towards the analytical solution._**
