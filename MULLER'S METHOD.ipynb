## Problem Statement: Use MÃ¼llerâ€™s method with guesses of ð‘¥0, ð‘¥1, and ð‘¥2 4.5, 5.5,and 5, respectively, to determine a root of the equation f(ð‘¥)= ð‘¥^3âˆ’ 13ð‘¥ âˆ’ 12

```python
import numpy as np
```

```python
def f(x):
    return x**3-13*x-12
```

```python
import pandas as pd
df=pd.DataFrame(columns=["X2","X1","X0","f(X2)","f(X1)","f(X0)","h1","Del1","h0","Del0","a","b","c","Det","X3","f(X3)","Error (%)"])
```

```python
def muller(x0,x1,x2,es):
    while True:
        h1=x2-x1
        h0=x1-x0
        del1=(f(x2)-f(x1))/(x2-x1)
        del0=(f(x1)-f(x0))/(x1-x0)
        a=(del1-del0)/(h1+h0)
        b=(a*h1)+del1
        c=f(x2)
        det=np.sqrt((b**2)-(4*a*c))
        if b>=0:
            x3=x2+((-2*c)/(b+det))
        elif b<0:
            x3=x2+((-2*c)/(b-det))
        if f(x3)==0:
            print(f"Root:{x3}")
            break
        ea=(abs((x3-x2)/x3))*100
        df.loc[len(df)]=[x2,x1,x0,f(x2),f(x1),f(x0),h1,del1,h0,del0,a,b,c,det,x3,f(x3),ea]
        x0=x1
        x1=x2
        x2=x3
        if ea<=es:
            print(f"Root:{x3}")
            break
```

```python
es=.02 #error tolerance 0.02%
#initial guesses
x0=4.5
x1=5.5
x2=5
muller(x0,x1,x2,es)
```
