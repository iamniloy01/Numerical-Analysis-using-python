**Given equation: 12x +3y -5z=1,
                  x +5y +3z =28 ,
                  3x +7y +13z =76 .
Use Jacobi method to find the solution. (Error tolerance 0.5%)initial guess x=1 , y=0 , z=1**


```python
import numpy as np
```

```python
a=np.matrix([
    [12,3,-5],
    [1,5,3],
    [3,7,13]
],dtype="float")
b=np.matrix([
    [1],
    [28],
    [76]
],dtype="float")
```

```python
import pandas as pd
df=pd.DataFrame(columns=["iteration","x","err(x)%","y","err(y)%","z","err(z)%"])
```

```python
ini=[1.0,0.0,1.0]
c=0
while True:
  copy=ini.copy()
  er=[]
  for i in range(len(a)):
    s=0
    for j in range(len(a)):
      if i!=j:
        s+=ini[j]*a[i,j]
    x=(b[i,0]-s)/a[i,i]
    copy[i]=x
    err=abs((x-ini[i]) / x)*100
    er.append(err)
  c+=1
  df.loc[len(df)]=[c,copy[0],er[0],copy[1],er[1],copy[2],er[2]]
  if all(e<=.5 for e in er):
    print("Solutions:",copy)
    break
  ini=copy.copy()
```

```python
print(df.to_string(index=False))
```
