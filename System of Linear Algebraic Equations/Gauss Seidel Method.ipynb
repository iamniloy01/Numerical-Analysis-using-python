**Given equation: 12x +3y -5z=1,
                  x +5y +3z =28 ,
                  3x +7y +13z =76 .
Use Gauss Seidel method to find the solution. (Error tolerance 0.5%)initial guess x=1 , y=0 , z=1**

```python
import numpy as np
```

```python
a=np.matrix([
    [12,3,-5],
    [1,5,3],
    [3,7,13]
],dtype="float")
b=np.matrix([
    [1],
    [28],
    [76]
],dtype="float")
```

```python
import pandas as pd
df=pd.DataFrame(columns=["iteration"," x "," err(x)% "," y "," err(y)% "," z "," err(z)% "])
```

```python
ini=[1.0,0.0,1.0]
c=0
while True:
  er=[]
  for i in range(len(a)):
    s=0
    for j in range(len(a)):
      if i!=j:
        s+=ini[j]*a[i,j]
    x=(b[i,0]-s)/a[i,i]
    err=abs((x-ini[i]) / x)*100
    er.append(err)
    ini[i]=x
  c+=1
  df.loc[len(df)]=[c,ini[0],er[0],ini[1],er[1],ini[2],er[2]]
  if all(e<=.5 for e in er):
    print("Solutions:",ini)
    break
```

```python
print(df.to_string(index=False))
```

**4X4 equation solve,
Given equation: 3x-y-t=-3,
                -x+3y-z=2,
                -y+4z-t=6,
                -x-z+4t=12
 initial value x=y=z=t=0, error tolerance 0.5%**

```python
A=np.matrix([
    [3,-1,0,-1],
    [-1,3,-1,0],
    [0,-1,4,-1],
    [-1,0,-1,4]
  ]  ,dtype="float"
)
B=np.matrix([
    [-3],
    [2],
    [6],
    [12]
],dtype="float")
```

```python
import pandas as pd
df=pd.DataFrame(columns=["iteration"," x "," err(x)% "," y "," err(y)% "," z "," err(z)% "," t "," err(t)% "])
```

```python
ini=[0.0,0.0,0.0,0.0]
c=0
while True:
  er=[]
  for i in range(len(A)):
    s=0
    for j in range(len(A)):
      if i!=j:
        s+=ini[j]*A[i,j]
    x=(B[i,0]-s)/A[i,i]
    err=abs((x-ini[i]) / x)*100
    er.append(err)
    ini[i]=x
  c+=1
  df.loc[len(df)]=[c,ini[0],er[0],ini[1],er[1],ini[2],er[2],ini[3],er[3]]
  if all(e<=.5 for e in er):
    print("Solutions:",ini)
    break
```

```python
print(df.to_string(index=False))
```
